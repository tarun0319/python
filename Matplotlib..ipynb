# MatplotLib

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
```

```python
x=[1,2,3,4,5,6,7]           
y=[78,85,52,45,56,32,88] 
```

```python
plt.plot(x,y,color='red')
```

```python
plt.plot(x,y,marker="o")
```

```python
plt.plot(x,y,marker="+")
```

```python
plt.plot(x,y,marker="*",markersize=8)
```

```python
plt.plot(x,y,marker="o",markersize=10,linestyle="-")
```

```python
plt.plot(x,y,marker="o",markersize=10,linestyle=":",linewidth=3)
```

```python
plt.plot(x,y,marker="o",markersize=10,linestyle=":",linewidth=3)
plt.xlabel('dayes',color='g')
plt.ylabel('spend_money',color='y')
plt.title('weekly spend money',color='r')
```

```python
## Creating Subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

```python
overs=[15,35,40,45,50]
ind=[60,80,115,200,215]
end=[45,90,114,210,218]
```

```python
plt.plot(overs,ind,color='red',marker='o',markersize=10,linestyle=':',linewidth=4,label='india')
plt.plot(overs,end,color='blue',marker='o',markersize=10,linestyle=':',linewidth=4,label='englad')
plt.xlabel('overs')
plt.ylabel('runs')
plt.title('today match')
plt.legend()  
plt.show()
```

```python
# Compute the x and y coordinates for points on a sine curve 
x = np.arange(0, 4 * np.pi, 0.1) 
y = np.sin(x) 
plt.title("sine wave form") 

# Plot the points using matplotlib 
plt.plot(x, y) 
plt.show() 
```

```python
## Bar plot

x = [2,8,10] 
y = [11,16,9]  

x2 = [3,9,11] 
y2 = [6,15,7] 
plt.bar(x, y) 
plt.bar(x2, y2, color = 'g') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  

plt.show()
```

```python
# Box Plot using Matplotlib
```

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=False);  
```

```python
# Pie Chart
```

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice
```

```python
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%2.2f%%' ,shadow=True)
plt.show()
```
